```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome(executable_path=r'C:\Users\Ihtesham\Downloads\Compressed\chromedriver_win32\chromedriver.exe')

driver.get("https://www.bestbuy.com/site/computer-accessories/hard-drives/abcat0504001.c?id=abcat0504001")

list = pd.DataFrame(columns = ['Product','Model','SKU', 'CurrentPrice', 'WasPrice']) 


country_selection = driver.find_element_by_class_name("country-selection")
united_states = country_selection.find_element_by_xpath("a[2]")
united_states.click()

pages = driver.find_elements_by_class_name('page-item')

for page in pages:
    pg = page.find_element_by_xpath("//a")
    print(pg.text)
    new_page = pg.get_attribute('href')
    driver.get(new_page)

    try:
        list_of_items = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CLASS_NAME, "sku-item-list"))
        )
    
        items = list_of_items.find_elements_by_class_name("sku-item")
    
        for item in items:
            header = item.find_element_by_class_name("sku-header")
            a_tag = header.find_element_by_tag_name("a")
            Product = a_tag.text
            print(Product)
        
            detail = item.find_element_by_class_name("sku-model")
        
            mod = item.find_element_by_class_name("information")
            model = detail.find_element_by_xpath("//div[2]/div[2]/div[1]/span[2]")
            Model = model.text
            #print('model',model.text)
        
            sku = detail.find_element_by_xpath("//div[2]/span[2]")
            SKU = sku.text
            #print('sku',sku.text)
        
            current_price = item.find_element_by_class_name("sku-list-item-price")
            price = current_price.find_element_by_xpath("//div/div/div[1]/div/div/div/div[2]/div/div[1]/div/span")
            #price = current_price.find_element_by_xpath("//div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/span")
            CurrentPrice = price.text
            #print(price.text)
        
            was_price = item.find_element_by_class_name("price-block")
            wasprice = was_price.find_element_by_xpath("//div[1]/div/div/div/div/div/div[1]/div[2]/div[1]/div[2]/span/div[1]")
            WasPrice = wasprice.text
            #print(wasprice.text)
        
            list.loc[len(list)] = [Product,Model,SKU,CurrentPrice,WasPrice]
    
    finally:
        driver.quit()

list
```
