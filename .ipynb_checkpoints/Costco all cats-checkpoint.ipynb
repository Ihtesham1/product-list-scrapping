```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome(executable_path=r'C:\Users\Ihtesham\Downloads\Compressed\chromedriver_win32\chromedriver.exe')

allCatLinks = []

driver.get("https://www.costco.com/SiteMapDisplayView")
site_map = driver.find_element_by_class_name("costcoBD-sitemap")
site_map = site_map.find_element_by_class_name("sitemap-section")
divs = site_map.find_elements_by_tag_name("div")
i = 0
for div in divs: 
    if i != 0:
        cols = div.find_elements_by_class_name("visible-xl")
        for col in cols:
            a_tags = col.find_elements_by_class_name("body-copy-link")
            for a_tag in a_tags:
                print("getting link of ", a_tag.text)
                href = a_tag.get_attribute("href")
                allCatLinks.append(href)
      
    else:
        i += 1

allCatLinks
```

```python
allItemLinks = []
driver = webdriver.Chrome(executable_path=r'C:\Users\Ihtesham\Downloads\Compressed\chromedriver_win32\chromedriver.exe')

def scrapper(link):

    driver.get(link)

    productList = driver.find_element_by_class_name("product-list")
    productList = productList.find_elements_by_class_name("product")
    for product in productList:
        description = product.find_element_by_class_name("description")
        a_tag = description.find_element_by_tag_name("a")
        print(a_tag.text)
        href = a_tag.get_attribute("href")
        allItemLinks.append(href)
        
        
for singleCatLink in allCatLinks:
    scrapper(singleCatLink)
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Users\Ihtesham\Downloads\Compressed\chromedriver_win32\chromedriver.exe')
list = pd.DataFrame(columns = ['Name','Price', 'Item_No', 'Model_No','Product_URL', 'Img_URL']) 

def getDetails(link):
    driver.get(link)
    productImageContainer = driver.find_element_by_id("productImageContainer")
    img = productImageContainer.find_element_by_id("productImage")
    src = img.get_attribute("src")
    Img_URL = src

    product_details = driver.find_element_by_id("product-details")
    h1 = product_details.find_element_by_xpath("//div[1]/div/div[1]/h1")
    Name = h1.text

    item_no = product_details.find_element_by_xpath("//div[1]/div/p/span")
    Item_No = item_no.text

    if product_details.find_element_by_xpath("//div[1]/div/p[2]/span"):
        model_no = product_details.find_element_by_xpath("//div[1]/div/p[2]/span")
        Model_No = model_no.text

    price = product_details.find_element_by_id("pull-right-price")
    price = price.find_element_by_class_name("value")
    Price = price.text
    
    Product_URL = link
    
    list.loc[len(list)] = [Name,Price, Item_No, Model_No,Product_URL, Img_URL]


getDetails("https://www.costco.com/teetot-pink-petal-fairy-child-costume.product.100676254.html")
    
list
```
