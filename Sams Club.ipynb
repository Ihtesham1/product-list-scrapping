```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome(executable_path=r'C:\Users\Ihtesham\Downloads\Compressed\chromedriver_win32\chromedriver.exe')

list = pd.DataFrame(columns = ['Book','Price']) 

pages = ["https://www.samsclub.com/b/books/15740371?clubId=4961&offset=0&searchCategoryId=15740371&selectedFilter=all&sortKey=relevance&sortOrder=1",
        "https://www.samsclub.com/b/books/15740371?clubId=4961&offset=48&searchCategoryId=15740371&selectedFilter=all&sortKey=relevance&sortOrder=1",
        "https://www.samsclub.com/b/books/15740371?clubId=4961&offset=96&searchCategoryId=15740371&selectedFilter=all&sortKey=relevance&sortOrder=1",
        "https://www.samsclub.com/b/books/15740371?clubId=4961&offset=144&searchCategoryId=15740371&selectedFilter=all&sortKey=relevance&sortOrder=1"]

for page in pages:

    driver.get(page)
    print(page)

    prod = WebDriverWait(driver, 50).until(
        EC.presence_of_element_located((By.CLASS_NAME, "sc-product-card"))
        )
    cards = driver.find_elements_by_class_name("sc-product-card")
    
    for card in cards:
        title = card.find_element_by_tag_name("h2")
        Book = title.text
        
        price_main = card.find_element_by_class_name("Price-characteristic")
        Price = price_main.text
        
        list.loc[len(list)] = [Book,Price]

    
list
```
